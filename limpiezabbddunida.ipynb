En "Salary" detectamos 102.260 valores nulos. Es una columna clave para el análisis por lo que vamos a imputarla de forma que les colocaremos la media salarial.

```python
media_salario = df["Salary"].mean()
df["Salary"].fillna(media_salario, inplace=True)
```

Las columnas "Cancellation Year" y "Cancellation Month" indican si un cliente ha cancelado su participación en el programa de fidelización. Sin embargo, estas columnas pueden contener valores nulos (NaN), lo cual no representa un error, sino que el cliente sigue activo.

Para simplificar, creamos una nueva columna llamada "Status_Client" con dos posibles valores:

"Active": cuando el cliente continua suscrito al programa (es decir, "Cancellation Year" es nulo).

"Cancelled": cuando el cliente ha cancelado su suscripción (es decir, "Cancellation Year" no es nulo).

```python
df["Client Status"] = df["Cancellation Year"].isna().map({ 
    True: "Active",
    False: "Cancelled"
})

df["Client Status"].value_counts()
```
